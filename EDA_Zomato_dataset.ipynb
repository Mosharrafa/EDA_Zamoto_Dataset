```python
Ratings=df1.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating count'})
```

```python
Ratings.head()
```

## Observation
1.Ratings in between 4.5 to 4.9 are excellent
2.Ratings in between 4.0 to 4.4 are very good
3.Ratings in between 3.5 to 3.9 are good
4.Ratings in between 3.0 to 3.4 are average
5.Ratings in between 2.5 to 2.9 are average
6.Ratings in between 2.0 to 2.4 are average

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,7)
sns.barplot(x='Aggregate rating',y='Rating count',data=Ratings)
```

```python
sns.barplot(x='Aggregate rating',y='Rating count',hue='Rating color',data=Ratings,palette=['blue','red','orange','yellow','green','green'])
```

```python
# Count plot
sns.countplot(x='Rating color',data=Ratings,palette=['blue','red','orange','yellow','green','green'])
```
